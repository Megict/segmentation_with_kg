```python
from docx import Document

document = Document("Глоссарий  ТУ.docx")
```

```python
fullText = []
for para in document.paragraphs:
    if para.text == 'АНГЛОЯЗЫЧНЫЕ ТЕРМИНЫ':
        break
    if (para.text.strip() != ''):
        fullText.append(para.text)
```

```python
terms = []
definitions = []
for para in fullText:
    print(para)
    print()
    print()
    if para.find('–') != -1:
        terms.append(para[0:para.find('–')].strip())
        definitions.append(para[para.find('–') + 1 :].strip())
```

```python
terms_definitions_dict = {}
for i in range(len(terms)):
    terms_definitions_dict[terms[i]] = definitions[i]
```
